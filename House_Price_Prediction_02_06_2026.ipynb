```python
#call alll the required library
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from google.colab import files
import pandas as pd
import zipfile

uploaded = files.upload()

# Get the name of the uploaded zip file
zip_filename = list(uploaded.keys())[0]

with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall()

# Replace with the actual CSV filename inside the zip, which appears to be 'data.csv'
df = pd.read_csv('data.csv')

df.head(10)
```

```python
# get all the column name from any data set and remove extra spc and make them all lower case
df.columns=df.columns.str.strip().str.lower()
print(df.columns)
```

```python
# convert all yes/no to 1/0 removing all the categorical values and add boolean values
# insted of using categorical values
# The original list contained an empty string and 'bedroom' which are not valid column names,
# leading to KeyError. Additionally, 'country', 'price', and 'view' are not binary 'yes'/'no' columns.
# Applying a 'yes'/'no' map to them would result in NaN values, which is likely not intended.
# For now, this list is empty as no suitable binary categorical columns were found.
binary_cols =[]
for i in binary_cols:
  df[i]=df[i].map({"yes":1,"no":0})
print(df.head())
```

```python
# SPLITING MY DATA SO NOW SELECT FEATURES --- IN TO TWO  ASIST FOR MODEL SETTING
x=df[['bedrooms','bathrooms','floors']]
y=df["price"]
```

full data(100) training 80% test 20%

```python
# here we're spliting our data set into two parts : traing testing (tp check the performence of the model)
# x-- it is variable it is independent variable
#y-- is dependent variable
#x_trai =80% traing || x_test --- testting 20%
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)

```

```python
 # train our model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
#predict
y_pred=model.predict(x_test)
#evaluate
print("MAE",mean_squared_error(y_test,y_pred))
print("R2",r2_score(y_test,y_pred))
```
